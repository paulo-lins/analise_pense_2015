### **PeNSE - Pesquisa Nacional de Saúde do Escolar**

## Com este lançamento, o IBGE apresenta comentários analíticos sobre os resultados da Pesquisa Nacional de Saúde do Escolar - PeNSE, realizada, em 2015, a partir de convênio celebrado com o Ministério da Saúde e o apoio do Ministério da Educação. A pesquisa, efetuada em consonância com as normas e diretrizes utilizadas em âmbito internacional e nacional para levantamentos envolvendo sujeitos humanos, em particular, adolescentes, teve por objetivo conhecer e dimensionar os diversos fatores de risco e de proteção à saúde desse grupo populacional.

## Nesta terceira edição da pesquisa, a ampliação de sua abrangência geográfica permitiu a desagregação dos dados por Unidades da Federação, além dos níveis já apresentados em edições anteriores. Outras inovações importantes foram introduzidas, dentre as quais se destaca a disponibilização de informações oriundas de dois planos amostrais distintos, que contemplam, respectivamente, escolares frequentando o 9o ano do ensino fundamental, e escolares de 13 a 17 anos de idade frequentando as etapas do 6o ao 9o ano do ensino fundamental (antigas 5a a 8a séries) e da 1a a 3a série do ensino médio, no ano de referência da pesquisa. A primeira amostra, tradicional da PeNSE, permite a comparação temporal entre os Municípios das Capitais e o Distrito Federal nas três edições da pesquisa – 2009, 2012 e 2015. A segunda amostra surge da demanda por dados passíveis de comparação com indicadores nacionais e internacionais sobre adolescentes e possibilita, entre outros aspectos, melhor identificação e acompanhamento de fatores relacionados ao desenvolvimento físico-biológico e ao tempo de exposição às condições de risco para esse grupo.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/paulo-lins/analise_pense_2015/main/PeNSE_2015_AMOSTRA2/arquivos%20csv/PENSE_AMOSTRA2_ALUNO.CSV'

dados = pd.read_csv(url, encoding='ISO-8859-1', sep=';', thousands='.', decimal=',')
dados
```

```python
dados.info()
```

```python
# observe que temos 3 colunas do tipo float64 vamos identificá-las
dados.select_dtypes(include='float64')
```

# Vamos fazer uma **tabela de frequência** para analisar a respostas de acordo com o sexo, masculino ou feminino

Quando um conjunto de dados passa a ser analisado na estatística, é importante levar em consideração suas diferentes categorias ou classes. Sendo assim, para cada uma delas, é preciso calcular uma frequência, conciliando uma distribuição. É por isso que a Tabela de Frequência se mostra relevante. fonte(https://blog.betrybe.com/tabela-de-frequencia/#1)

```python
# Vamos começar pela pergunta 'VB01001' que é 'Qual o seu sexo?', 
# com respostas 1 (masculino) e 2 (feminino)
# abaixo teremos a frequÊncia relacionada ao sexo dos estudantes

dados['VB01001'].value_counts()
```

```python
# utilizaremos o parâmetro 'normalize' e multiplicaremos por 100
# para ver o resultado em valores percentuais
dados['VB01001'].value_counts(normalize = True) * 100
```

# Observamos que o número de alunos e alunas que respondeu à pesquisa é praticamente o mesmo, o que, provavelmente faz parte da estratégia de amostragem da pesquisa

Em alguns casos teremos que **normalizar** as variáveis

**Normalização** são proporções sem unidades de medida (adimensionais ou invariantes de escala) que nos permitem comparar elementos de diferentes variáveis ​​e diferentes unidades de medida.

```python
# Vamos calcular a frequência relacionada à pergunta VB11007 'Como você se sente em relação ao seu corpo?'
frequencia = dados['VB11007'].value_counts()
frequencia
```

```python
# Vamos calcular o percentual relacionado à pergunta VB11007 'Como você se sente em relação ao seu corpo?'
percentual = dados['VB11007'].value_counts(normalize = True) * 100
percentual
```

```python
# vamos unificar essas Series e fazer uma tabela
respostas = {
    1  : 'Muito satisfeito(a)',
    2  : 'Satisfeito(a)',
    3  : 'Indiferente',
    4  : 'Insatisfeito(a)',
    5  : 'Muito insatisfeito(a)',
    99  : 'Não informado'
}

dist_freq = pd.DataFrame({'Frequência': frequencia, 'Percentual': percentual})
dist_freq
```

```python
# vamos analisar o IMC dos estudantes e classificá-los
# para isso, usaremos as variáveis peso e altura, respectivamente 'VB17001' 
# e 'VB17002'. A fórmula do imc é: IMC = Peso ÷ (Altura × Altura)

dados1 = dados
```

## **Vamos converter uma variável contínua em uma categórica utilizando as classificaçoes do IMC e a função cut do pandas**

Use cut quando precisar segmentar e classificar valores de dados em compartimentos. Esta função também é útil para passar de uma variável contínua para uma variável categórica. Por exemplo, cut pode converter idades em grupos de faixas etárias. Suporta binning em um número igual de bins, ou uma matriz pré-especificada de bins.

```python
dados1['IMC'] = dados['VB17003'] / ((dados['VB17004']/100) ** 2)
```

```python
dados1[['VB17003', 'VB17004', 'IMC']].head()
```

```python
# Agora vamos fazer a classificação conforme o IMC
classes_imc = [0, 18.5, 25, 30, 60]
labels_imc = ['abaixo', 'normal', 'sobrepeso', 'obesidade']

#vamos utilizar o cut para fazer a classificação
dados1['classe_IMC'] = pd.cut(x = dados1['IMC'],
       bins = classes_imc,
       labels = labels_imc,
       include_lowest = True)
```

```python
dados1.head()
```

```python
frequencia = dados1['classe_IMC'].value_counts()
percentual = dados1['classe_IMC'].value_counts(normalize = True) * 100
dist_freq_imc = pd.DataFrame({'Frequência': frequencia, 'Percentual': percentual})
dist_freq_imc
```

```python
dist_freq_imc.plot(figsize=(10,10),
                   kind='bar')
```
